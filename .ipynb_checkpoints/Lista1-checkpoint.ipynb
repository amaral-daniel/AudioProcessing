```python
import matplotlib.pyplot as plt
from scipy import signal
from scipy.io import wavfile
import os
import IPython
import math
```

```python
base_dir="/Users/danielmendonca/git/AudioProcessing/AudioFiles"
file_1="Brandenburg Concerto No. 3 In G Major- I-Allegro_RED.wav"
file_2="Two Part Invention In F Major_RED.wav"
file_3="When I am laid in earth (Dido_s lament)_RED.wav"
file_4="h√° gente aqui_RED.wav"
```

#### Brandenburg Concerto No. 3 In G Major- I-Allegro

```python
file_path = os.path.join(base_dir,file_1)
sample_rate,samples = wavfile.read(file_path)
IPython.display.Audio(file_path)
```

Since the variation sounds fast, we set the window length to 20ms, which is the minimum variation we (humans) can hear

```python
window_length=0.02 #20ms, minimum variation we (humans) can hear
nfft=int(window_length*sample_rate)
nfft
```

```python
plt.specgram(samples, Fs=sample_rate,cmap='jet_r',NFFT=nfft,noverlap=nfft/2, title= 'Spectrogram',x_label='time (s)', y_label='frequency (Hz)')
plt.colorbar()
plt.show()
```
